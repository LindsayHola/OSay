```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def carprobl(Orlando,Tampa):
    O=Orlando
    T=Tampa
    oarr = []
    tarr = []
    for i in range(10):
        Onext = 0.6*O + 0.3*T
        oarr.append(Onext)
        Tnext = 0.4*O+ 0.7*T
        tarr.append(Tnext)
        T = Tnext
        O = Onext
    return [oarr, tarr]
```

```python
oarr, tarr = carprobl(7000,0)
x = [i for i in range(10)]
y = oarr
plt.plot(x,y, label = "Orlando")
plt.plot(x,tarr, label = "Tampa")
plt.legend()
plt.title("Rental Car problem, Orlando starts with 7000, Tampan starts with 0");
```
